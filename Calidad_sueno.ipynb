Analisis de Edades

```python
#cantidad de hombres y mujeres
df['Gender'].value_counts()
```

```python
#promedio de edades del conjunto
df['Age'].mean()
```

```python
df['Age'].std()
#el conjunto de datos, se inclina a un lado
```

```python
#segmentar en adultos y adultos viejos
df.loc[df['Age']<=40,'Age_group']='Adult'
df.loc[df['Age']>=41,'Age_group']='Older Adult'
```

```python
df.head()
```

```python
#conteo de adultos
df['Age_group'].value_counts()
#existen menos adultos, el dataset tiene mas adultos mayores
```

```python
#agrupar por genero
df.groupby('Gender')['Age'].agg(['mean','std'])
#obtener el promedio y desviacion por genero
```

Nuestra poblacion tiende a ser mayoritariamente mayor en edad, especialmente en la poblacion de mujeres, por lo que hacer un analisis por medio de la edad para determinar la calidad y tiempo de sueño, podria estar sesgada por la edad y los habitos de las personas mayores

Analisis por ocupacion

```python
#conocer las ocupaciones por genero
df.groupby('Occupation')['Gender'].value_counts()
```

```python
#tiempo maximo de sueño
print(df['Sleep Duration'].max())
```

```python
#calidad de sueño por genero
df.groupby('Gender')[['Sleep Duration','Quality of Sleep']].mean()
```

Tanto hombres como mujeres duermen una cantidad similar de horas, y las recomendadas por la OMS, podemos observar que las mujeres tiene una ligera mejor calidad del sueño

```python
#calidad de sueño por ocupacion
occ_qua = df.groupby('Occupation')[['Sleep Duration','Quality of Sleep']].mean()
occ_qua.sort_values(by='Quality of Sleep')
```

Debido a la distribucion no equitativa de profesiones en los encuestados, es dificil determinar que profesion presenta un mejor calidad de sueño, pero podemos observa que conforme mas tiempo de sueño se tenga mejor calidad de sueño tendremos. Podemos observar que las profesiones con mejor sueño son Ingenieros, abogados y contadores. Mientras que la peor calidad de sueño la tienen los vendedores

```python
#indice de Masa corporal por ocupacion
occ_bmi = df.groupby('Occupation')['BMI Category'].value_counts()
occ_bmi.sort_values()
```

```python
#Nivel de estres por ocupacion
occ_stress = df.groupby('Occupation')['Stress Level'].value_counts()
occ_stress.sort_values()
```

Las personas con mayor indice de masa corporal son enfermeras, mientra que la profesion con mayor nivel de estres son los doctores

Analisis por actividad fisica

```python
#Calidad de sueño, paso por ocupacion
occ_steps = df.groupby('Occupation')['Daily Steps','Physical Activity Level','Sleep Duration','Quality of Sleep'].mean()
occ_steps.sort_values(by='Daily Steps')
```

Las personas con mas actividad fisica y mayor cantidad de pasos por dia son enfermeras, lo que es esperado debido a la implicacion de las actividades que realizan en esta profesion.

Por su parte, la cantidad de pasos de una persona no esta extrictamente ligada con un tiempo de actividad similar a la cantidad de pasos, por lo que si analizamos la cantidad de paso con respecto al tiempo y calidad podemos, observar que no exites una relacion proporcional entre estos.

```python
occ_steps.sort_values(by='Physical Activity Level')
```

Mientras que la cantidad de actividad fisica realizada en el dia, tiene una mejor relacion con respecto a la calidad de sueño de las personas, por lo que podriamos decir que entre mas cercana sea el tiempo de actividad a una hora por dia, entonces se tendra una mejor calidad de sueño y una mayor duracion de este, en niveles aceptables y recomendados

```python
bmi = df.groupby('BMI Category')[[ 'Heart Rate','Sleep Duration','Quality of Sleep']].mean()
bmi.sort_values(by='Quality of Sleep')
```

Podemos decir que estas variables si estan relacionadas, un indice de masa corporal normal presenta una mejor calidad y tiempo de sueño, y conforme este aumenta, el sueño disminuye en las personas. Por su parte conforme menor sea la frecuencia cardica mejor calidad y tiempo de sueño tendra la persona

```python
bmi_blood = df.groupby('Blood Pressure')[['Sleep Duration','Quality of Sleep']].value_counts()
bmi_blood
```

Conforme a los datos, podemos decir que no existe una relacion que demuestre que una presion normal (120/80), mejore la calidad o tiempo de sueño, ni que una presion mayor afecte la calidad de sueño
